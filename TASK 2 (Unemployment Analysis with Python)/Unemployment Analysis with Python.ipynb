```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('Unemployment in India.csv')

# Explore the dataset to understand its structure and missing values
print(data.head())
print(data.info())

```

```python
# Convert the 'Date' column to strings
data[' Date'] = data[' Date'].astype(str)

# Remove leading and trailing spaces from the 'Date' column
data[' Date'] = data[' Date'].str.strip()

# Convert the 'Date' column to a proper datetime format
# Convert the 'Date' column to a proper datetime format
data[' Date'] = pd.to_datetime(data[' Date'])


# Check for missing values
print(data.isnull().sum())

# If there are missing values, handle them appropriately, e.g., drop rows with missing values
data.dropna(inplace=True)
```

```python
# Perform exploratory data analysis (EDA)
# Example: Plotting the unemployment rate over time
plt.figure(figsize=(10, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=data)
plt.title(' Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel(' Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter data for Covid-19 period (assuming you have a specific date range)
start_date = '2020-03-01'
end_date = '2021-03-01'
covid_period_data = data[(data[' Date'] >= start_date) & (data[' Date'] <= end_date)]

# Calculate the average unemployment rate during Covid-19
average_unemployment_rate_covid = covid_period_data[' Estimated Unemployment Rate (%)'].mean()
print(f'Average Unemployment Rate during Covid-19: {average_unemployment_rate_covid:.2f}')
```
